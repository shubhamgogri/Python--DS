```python
import matplotlib.pyplot as plot
import numpy as np
```

```python
x = np.arange(0,9)
y = np.arange(11,20)
```

```python
#  scatter plot

plot.scatter(x,y, c='g')
plot.xlabel('X axis')
plot.ylabel('Y axis')
plot.title('Graph')
```

```python
y = x*x
```

```python
# plot

plot.plot(x,y, 'r', linestyle= 'dashed', linewidth = 2, markersize = 50)
plot.xlabel('X axis')
plot.ylabel('Y axis')
plot.title('Graph')

# plot.savefig("scater.png")
```

```python
# subplot
plot.subplot(2,2,1) 
# rows, column, pos
plot.plot(x,y)

plot.subplot(2,2,2)
plot.scatter(x,y, c='g')
plot.xlabel('X axis')
plot.ylabel('Y axis')
plot.title('Graph') 

plot.subplot(2,2,3)
plot.plot(x,y, c='g', linestyle = "dashed")
plot.xlabel('X axis')
plot.ylabel('Y axis')
plot.title('Graph')

plot.subplot(2,2,4)
plot.plot(x,y, 'r*-')
plot.xlabel('X axis')
plot.ylabel('Y axis')
plot.title('Graph')
```

```python
y = 3*x -10 + x*x
plot.plot(x,y, c='g')
plot.xlabel('X axis')
plot.ylabel('Y axis')
plot.title('Graph')
```

```python
# sine waves

x = np.arange(0,5*np.pi,0.1)
y = np.cos(x)
plot.plot(x,y, c='g')
plot.xlabel('X axis')
plot.ylabel('Y axis')
plot.title('Sine wave')
```

```python
# subplot sin and cos wave
y_sin = np.sin(x)
y_cos = np.cos(x)

plot.subplot(2,1,1)
plot.plot(x,y_sin, c='g')
plot.xlabel('X axis')
plot.ylabel('Y axis')
plot.title('Sine')

plot.subplot(2,1,2)
plot.plot(x,y_cos, c='g')
plot.xlabel('X axis')
plot.ylabel('Y axis')
plot.title('cos')
```

```python
# Bar plot
x = [1,2,3,4,5]
y = [5,2,4,4,6]
plot.bar(x,y)
```

```python
# HIstograms

# frequency of numbers in predefined buckets
a = np.array([55,15,13,51,55,10,35,20,44,1,35,24,26,43,48,30])
plot.hist(a)
```

```python
# Box plot
# helps in getting percentiles

data = [np.random.normal(0,std,100) for std in range(1,4)]
data
plot.boxplot(data, vert = True, patch_artist = True)
```

```python
# pie chart

labels =['Python', 'C++', 'Java', 'Ruby']
sizes = [210,130,250,210]
colors = ['gold','blue', 'green', 'red']
explode = (0.1,0,0,0)
plot.pie(sizes, explode = explode, labels = labels, colors = colors ,
         autopct = '%1.1f%%', shadow = True)
```
