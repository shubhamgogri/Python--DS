MultiCollinearity :

It is an issue that take place when there are two features that are corelated on each other for about more than 90%. 


```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df_sal = pd.read_html('https://github.com/krishnaik06/Multicollinearity/blob/master/data/Salary_Data.csv')[0]
```

```python
df_sal.drop(['Unnamed: 0'], inplace = True, axis = 1)
```

```python
X = df_ad[['TV', 'radio','newspaper']]
y = df_ad['sales']
```

Ordinary Least square

y = B0 + B1(TV) + B2 (Radio) + B3 (Newspaper)

Here B0 does not exist and we need to create that column 

```python
import statsmodels.api as sm
X = sm.add_constant(X)
```

```python
X
```

```python
# fit a OlS model with intercept on TV and Radio
model = sm.OLS(y,X).fit() 
```

```python
model.summary()
```

Coef values are Beta ones

NO Collinerity in features of the model

observe the Std error if the value is high then an issue is observed
else it is fine

Second way to identify

```python
X.iloc[:, 1:].corr()
```
