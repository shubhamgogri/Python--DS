```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
# Correlation with Heatmap
# get tips or assume it from the other features
# tip = Dependent Feature
#  others are independent feature
df.corr()
# only for int or float
```

```python
sns.heatmap(df.corr())
```

```python
# Joint Plot

# Bivariate 
sns.jointplot(x = 'tip', y = 'total_bill', data = df, kind = 'hex')
```

```python
# REgresion
sns.jointplot(x = 'tip', y = 'total_bill', data = df, kind = 'reg')
```

```python
# Pair plot when more than two features
sns.pairplot(df)
```

```python
# use other features apart from int and float
sns.pairplot(df, hue = 'sex')
```

```python
# Dist plot
sns.distplot(df['tip'])
```

Categorical Features

```python
 sns.countplot("day", data = df)
```

```python
sns.barplot(x = "day", y = 'total_bill', data = df)
```

```python
# box plot
sns.boxplot('total_bill', 'sex', data = df, hue = 'day')
```

```python
# violin plot

sns.violinplot(x = 'total_bill',y =  'day',data =  df)
```
